```python
import random
import satx

c = random.randint(-10, 10)

satx.engine(10)

x = satx.integer()
y = satx.integer()

assert x - y == c

while satx.satisfy():
    if x - y != c:
        print(c, x - y + satx.oo(), x, y, '*')
    else:
        print(c, x - y, x, y)
```

```python
import random
import satx

c = random.randint(-10, 10)

satx.engine(10)

x = satx.integer()
y = satx.integer()

assert x / y == c

while satx.satisfy():
    if x.value // y.value != c:
        print(c, (x.value + satx.oo()) // y.value, x, y, '*')
    else:
        print(c, x.value // y.value, x, y)
```
